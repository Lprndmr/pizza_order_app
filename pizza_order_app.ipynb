```python
# Gerekli kütüphaneleri ekleme:
import csv
from datetime import datetime

# Pizza üst sınıf oluşturulması:
class Pizza:
    def __init__(self):
        self.description = "Unknown Pizza"
        self.cost = 0.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

# Pizza alt sınıf oluşturulması:
class ClassicPizza(Pizza):
    def __init__(self):
        super().__init__()                                                        # Alt sınıfa ait özelliklerin yazılması(tanım ve fiyat):
        self.description = "Classic Pizza: This pizza does not have tomato sauce and instead uses a mixture of ricotta cheese, garlic, and olive oil as the base. It is then topped with mozzarella cheese and any other toppings of your choice, such as spinach or mushrooms."    
        self.cost = 10.0


class MargheritaPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Margherita Pizza: This classic Neapolitan pizza features a thin, crispy crust topped with tangy tomato sauce, fresh mozzarella cheese, and a few leaves of basil."
        self.cost = 12.0


class TurkPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Turk Pizza: Turkish Pizza, also known as Lahmacun, is a popular Middle Eastern dish that consists of a thin, round flatbread topped with a mixture of ground meat (usually lamb or beef), vegetables (such as onions and tomatoes), herbs (such as parsley and mint), and spices (such as paprika and cumin)."
        self.cost = 14.0


class DominosPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Dominos Pizza: Domino's Pizza is a popular American chain of pizza restaurants that offers a variety of pizza flavors, sides, and beverages. Their pizzas typically feature a hand-tossed or thin crust, tomato sauce, and mozzarella cheese, along with various toppings depending on the pizza flavor."
        self.cost = 16.0

# Decorator üst sınıf oluşturulması:
class Decorator(Pizza):
    def __init__(self, pizza):
        super().__init__()
        self.pizza = pizza

    def get_description(self):
        return self.pizza.get_description() + ", " + self.description

    def get_cost(self):
        return self.pizza.get_cost() + self.cost

# Decorator alt sınıf oluşturulması:
class Olives(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Olives"
        self.cost = 2.0


class Mushrooms(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Mushrooms"
        self.cost = 2.5


class GoatCheese(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Goat Cheese"
        self.cost = 3.0


class Meat(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Meat"
        self.cost = 4.0


class Onions(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Onions"
        self.cost = 1.5


class Corn(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.description = "Corn"
        self.cost = 1.0

# Menu.txt'i görüntüleme:
def display_menu():
    with open("Menu.txt", "r") as menu_file:
        print(menu_file.read())

# Kullanıcıya pizza tercihinde bulundurtma:
def get_pizza_choice():
    choice = input("Please choose a pizza base (1-4): ")
    if choice == "1":
        return ClassicPizza()
    elif choice == "2":
        return MargheritaPizza()
    elif choice == "3":
        return TurkPizza()
    elif choice == "4":
        return DominosPizza()
    else:
        print("Invalid choice. Please choose a number between 1 and 4.")
        return get_pizza_choice()

# Kullanıcıya sos tercihinde bulundurtma:
def get_sauce_choice(pizza):
    while True:
        sauce_choice = input("Please choose a sauce for your pizza (11-16) or press enter to finish: ")
        if sauce_choice == "":
            return pizza
        elif sauce_choice == "11":
            pizza = Olives(pizza)
        elif sauce_choice == "12":
            pizza = Mushrooms(pizza)
        elif sauce_choice == "13":
            pizza = GoatCheese(pizza)
        elif sauce_choice == "14":
            pizza = Meat(pizza)


def main():
    display_menu()

    # Pizza seçimlerini alma:
    pizza = get_pizza_choice()

    # Sos seçimlerini alma:
    pizza = get_sauce_choice(pizza)

    # Toplam ücreti hesaplama:
    total_cost = pizza.get_cost()
    print(f"Total cost of your order is {total_cost:.2f} TL.")

    # Kullanıcı bilgisi alma:
    name = input("Enter your name: ")
    id_num = int(input("Enter your ID number: "))
    cc_num = int(input("Enter your credit card number: "))
    cc_password = input("Enter your credit card password: ")

    # Verileri Orders_Database.csv'e kaydetme:
    with open("Orders_Database.csv", mode="a", newline="") as csv_file:
        writer = csv.writer(csv_file)
        now = datetime.now()
        dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
        writer.writerow([name, id_num, cc_num, cc_password, pizza.get_description(), total_cost, dt_string])

if __name__ == "__main__":

    main()
```
